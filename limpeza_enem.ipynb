```python
import pandas as pd

arquivo = 'RESULTADOS_2024.csv'

# Colunas que REALMENTE existem no seu arquivo, baseadas no seu print:
colunas_uteis = [
    'NU_SEQUENCIAL', 'SG_UF_ESC', 'NO_MUNICIPIO_ESC', 
    'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO'
]

print("Iniciando a filtragem... Aguarde um momento.")

# Lendo o arquivo (usando chunksize para garantir que não trave)
chunks = pd.read_csv(arquivo, sep=';', encoding='latin-1', usecols=colunas_uteis, chunksize=100000, low_memory=False)

lista_df = []

for chunk in chunks:
    # Removemos quem não tem nota de Matemática ou Redação (faltosos)
    chunk_limpo = chunk.dropna(subset=['NU_NOTA_MT', 'NU_NOTA_REDACAO'])
    lista_df.append(chunk_limpo)

# Junta tudo em um DataFrame final
df_final = pd.concat(lista_df)

# Salva o arquivo pronto para o Power BI
df_final.to_csv('enem_resultados_limpos.csv', index=False)

print(f"Sucesso! O arquivo 'enem_resultados_limpos.csv' foi gerado com {len(df_final)} linhas.")
```
